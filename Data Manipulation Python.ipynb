```python
#Load Metro_Interstate_Traffic_Volume.csv as DataFrame
df = pd.read_csv('/content/Metro_Interstate_Traffic_Volume.csv')
df.head(5)
```

```python
#Create ‘month’ column 
df['month'] = pd.to_datetime(df['date_time']).dt.month
df.sample(5)
```

```python
#Create ‘day_of_month’ column
df['date_time'] = pd.to_datetime(df['date_time'])
df['day_of_month'] = df['date_time'].dt.day
df.sample(5)
```

```python
#Create ‘day_name’ column
df['day_name'] = pd.to_datetime(df['date_time']).dt.day_name()
df.sample(5)
```

```python
#Create ‘hour’ column
df['hour'] = pd.to_datetime(df['date_time']).dt.hour
df.sample(5)
```

```python
#Create ‘is_weekend’ binary column
df['is_weekend'] = df['day_name'].apply(lambda x: 1 if (x == 'Saturday' or x =='Sunday') else 0)
df.sample(5)
```
